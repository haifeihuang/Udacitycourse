```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import nan as NA
import seaborn as sns
```

```python
data = pd.read_csv("titanic_data.csv", engine = 'python')
```

```python
data.info()
```

```python
survived = pd.DataFrame(data.Survived.value_counts())
```

```python
# The death rate of passengers on board
labelsl = 'Died','Survived'
plt.axes(aspect=1)
plt.pie(survived,labels = labelsl, autopct='%.0f%%')
plt.title('Survival Rate')
```

```python
# Group people by survived or not
survived_df = data.groupby('Survived')
sex_df = survived_df.Sex.value_counts()
sex_df
sex_df.unstack()
```

```python
# Age analysis
# Ratio of male and female die or survived in the accident
plt.figure(figsize = (12, 5))
plt.subplot(121)
label = 'Male Die','Female Die','Female Survived','Male Survived'
plt.axes(aspect = 1)
plt.pie(sex_df, labels = label, autopct = '%.0f%%')
plt.title('Survival Rate of Sex')
sex_df.unstack().plot(kind = 'bar')
```

```python
#From age analysis we can see that female has a lower death rate and higher survival rate than man, meaning that
#female has more chance to survive in the accident.
```

```python
# Pclass analysis
# Survival comparison of people form different classes
tmp =  survived_df.Pclass.value_counts()
a = tmp.unstack()
a.plot(kind = 'barh', stacked = True, alpha = 0.5)
plt.show()
```

```python
# From Pclass analysis we can see that people who are from class 3 are more likely to die than people from class 2 and 1.
```

```python
# Age & Pclass analysis
pclass_sex_df = data.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack().plot(kind='bar')
plt.title('Sex VS. Pclass VS. Survival Rate')
plt.ylabel('Survival Rate')
```

```python
# Whether the person is female or not, the survival rate rises when have lower Pclass number.
```

```python
# Age analysis
# Extract survived passengers for analysis
survived_only = data[data['Survived'] == 1]
died = data[data['Survived'] == 1]
first_clean = survived_only.drop('Cabin', axis = 1)
second_clean = first_clean.dropna()
```

```python
# Divide passengers according to their ages
teenager_sur = second_clean[second_clean['Age'] <=12].Age.count()
youth_sur = second_clean[(12 < second_clean['Age']) & ( second_clean['Age']<= 18 )].Age.count()
adult_sur = second_clean[(18 < second_clean['Age']) & ( second_clean['Age']<= 65 )].Age.count()
old_sur = second_clean[second_clean['Age'] > 65 ].Age.count()
age_df = pd.DataFrame([teenager_sur,youth_sur,adult_sur,old_sur], index = ['teenager','youth','adult','old'], columns = ['Number'])
```

```python
# Determine survival number of people and boxplot of survival trend
plt.figure(figsize= (12,5))
plt.subplot(121)
second_clean['Age'].hist(bins = 70)
plt.xlabel('Age')
plt.ylabel('Number')
plt.subplot(122)
second_clean.boxplot(column = 'Age', showfliers = False)
```

```python
# Further analysis using percentage and bar plot
plt.figure(figsize = (10,5))
age_df.plot(kind = 'bar', align = 'center', alpha = 0.5)
plt.xlabel('Age')
plt.ylabel('Number')
plt.title('Number of people of different age')
plt.show()
plt.subplot(122)
label = 'teenager','youth','adult','old'
plt.pie(age_df,labels = label, autopct='%.0f%%')
plt.title('Percentage of people of different age')
plt.show()
```

```python
#From graphs above, we can the mean age of survived people are lower 30, and audlts takes up majority of the total survial
#share.
```

```python
sibsp_df = data[data['SibSp'] != 0].groupby('Survived').SibSp.count()
plt.figure(figsize = (12,5))
plt.subplot(121)
sibsp_df.plot(kind = 'bar')
plt.xlabel(' Survived or not')
plt.ylabel('Number')
plt.title('Survival rate for people with siblings ')
plt.subplot(122)
nonsibsp_df = data[data['SibSp'] == 0].groupby('Survived').SibSp.value_counts()
nonsibsp_df.plot(kind = 'bar')
plt.xlabel(' Survived or not')
plt.ylabel('Number')
plt.title('Survival rate for people without siblings ')
```

```python
# From graphs above, we can see that people with siblings have higher survival rate than the others who not.
```

```python
# Parch analysis
sibsp_df = data[data['Parch'] != 0].groupby('Survived').Parch.count()
plt.figure(figsize = (10,5))
plt.subplot(121)
sibsp_df.plot(kind = 'bar')
plt.xlabel(' Survived or not')
plt.ylabel('Number')
plt.title('Survival rate for people with parents or children ')
plt.subplot(122)
nonsibsp_df = data[data['Parch'] == 0].groupby('Survived').Parch.value_counts()
nonsibsp_df.plot(kind = 'bar')
plt.xlabel(' Survived or not')
plt.ylabel('Number')
plt.title('Survival rate for people without parents or children ')
```

```python
# From Parch analysis, we can see that people with parents or children have a higher survival rate than others.
```

```python
# To conclude, people who are female, or who are from number of lower Pclass, or who has an age between 19 to 35, or who have siblings or parents and children,
# have higher survival rate than other ones.
```
